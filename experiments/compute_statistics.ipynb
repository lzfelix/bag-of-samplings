```python
from scripts import loader
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True)
```

```python
def load_data(filepath: str):
    X_hea = loader.load_fold_signals(filepath, fold_id=0, is_healthy=True, sampling_factor=1, verbose=0)
    X_pat = loader.load_fold_signals(filepath, fold_id=1, is_healthy=False, sampling_factor=1, verbose=0)

    df_hea = pd.concat([pd.DataFrame(sample) for sample in X_hea])
    df_pat = pd.concat([pd.DataFrame(sample) for sample in X_pat])

    df_all = pd.concat([df_hea, df_pat])

    channel_descriptions = list()
    for channel in range(df_all.shape[-1]):
        description = df_all[channel].describe([.01, 0.9])
        channel_descriptions.append(description)

    header_names = [f'Channel #{index}' for index in range(6)]
    channel_descriptions = pd.concat(channel_descriptions, axis=1, keys=header_names)
    return channel_descriptions
```
